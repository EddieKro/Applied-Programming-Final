```python
data = pd.read_csv('/kaggle/input/covid19-symptoms-checker/Cleaned-Data.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import RidgeClassifier
from sklearn.metrics import mean_squared_error,f1_score,fbeta_score

x_train,x_test,y_train, y_test = train_test_split(data,data.Infected)
clf = RidgeClassifier().fit(x_train,y_train)
pred = clf.predict(x_test)
fbeta_score(y_test,pred,0.4)
```

```python
def process_question(ans,answers):
    ans_values = [0 for i in range(len(answers))]
    
    for i in range(len(answers)):
        if ans==answers[i]:
            ans_values[i]=1
            break
    
    return ans_values
```

```python
process_question('y',['dn','n','y'])
```

```python
def age_group(age):
    age_group_values = [0 for i in range(5)]
    age_groups = [10,20,25,60,61]
    for i in range(len(age_groups)):
        if age<age_groups[i]:
            break
    age_group_values[i]=1
    return age_group_values
```
